```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import *
```

```python
df = pd.read_csv('data/data_v5.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0.1','Unnamed: 0','Year','t1_TEAM','t2_TEAM'],inplace=True)
```

```python
X = df.drop(columns=['outcome']).to_numpy()
y = df['outcome'].to_numpy()
mi = mutual_info_classif(X,y)
```

```python
k_best = SelectKBest(mutual_info_classif,k=25).fit(X,y)
```

```python
k_best
```

```python
selected_fts = k_best.get_feature_names_out(list(df.columns).remove('outcome'))
```

```python
fts_list = []
for s in selected_fts:
    fts_list.append(int(s[1:]))
```

```python
subset_df = df.iloc[:,fts_list]
```

```python
subset_df['outcome']=y
```

```python
subset_df.head()
```

```python
subset_df.to_csv('data/data_v6_k25.csv')
```
