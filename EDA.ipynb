```python
cols = df.columns
idx = 0
for col in cols:
    if 't1' in col or col == 'Year':
        idx+=1
    else:
        break 
    
```

```python
idx
```

```python
t1_data = df.iloc[:,:idx] #data associated with t1 + year
```

```python
t2_data = df.iloc[:,idx:] #data associated with t2 + outcome
```

```python
t1_data.head()
```

```python
t2_data.head()
```

```python
t1_data.drop(columns=['Year','t1_TEAM','t1_seed','t1_series_wins'],inplace=True)
t2_data.drop(columns=['t2_TEAM','t2_seed','t2_series_wins','outcome'],inplace=True)
```

```python
len(t1_data.columns)==len(t2_data.columns) #same number of cols verification
```

```python
t1_cols = list(t1_data.columns)
t2_cols = list(t2_data.columns)
suffixes = []
for i in range(len(t1_data.columns)):
    t1_col = t1_cols[i]
    t1_col_suffix = t1_col.replace('t1_','')
    suffixes.append(t1_col_suffix)
    t2_col = t2_cols[i]
    if 't2_'+t1_col_suffix != t2_col:
        print(i) #verifying that columns capture identical statistic
```

```python
import numpy as np
```

```python
t1_arr = t1_data.to_numpy()
t2_arr = t2_data.to_numpy()
```

```python
arr = np.vstack((t1_arr,t2_arr))
corr_df = pd.DataFrame(arr,columns=suffixes)
corr_df.head()
```

```python
corr_df.corr()
```
