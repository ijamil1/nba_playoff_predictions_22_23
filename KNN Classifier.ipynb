```python
from sklearn import neighbors, datasets
```

```python
import pandas as pd
```

```python
data = pd.read_csv('data/data_v4.csv')
data.head()
```

```python
#no need for Year or team names
data.drop(columns=['Unnamed: 0','Year','t1_TEAM','t2_TEAM'],inplace=True)
data.head()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
#split data into train and test split

X = data.drop(columns=['outcome'])
X = X.to_numpy()
y = data['outcome'].to_numpy()
x_train, x_test, y_train, y_test = train_test_split(X,y)
```

```python
#fit Vanilla knn classifier

knn = neighbors.KNeighborsClassifier()
knn.fit(x_train,y_train)
```

```python
y_preds = knn.predict(x_test)
```

```python
def accuracy(a1, a2):
    n = a1.shape[0]
    correct = 0
    for i in range(n):
        if a1[i]==a2[i]:
            correct+=1
    return correct/n
```

```python
accuracy(y_preds,y_test) #poor accuracy
```

```python
#let weight of neighbors be inversely proportional to distance from point when it comes to weighting the 'votes'
knn1 = neighbors.KNeighborsClassifier(weights='distance')
knn1.fit(x_train,y_train)
y_preds = knn1.predict(x_test)
accuracy(y_preds,y_test)
```

```python
#test what the best value for the hyperparameter controlling the number of neighbors to consider during classification is

res = {}

for k in range(3,16):
    res[k] = []
    for i in range(100):
        x_train, x_test, y_train, y_test = train_test_split(X,y)
        knn = neighbors.KNeighborsClassifier(n_neighbors=k)
        knn.fit(x_train,y_train)
        ypreds = knn.predict(x_test)
        res[k].append(accuracy(ypreds,y_test))

for k in range(3,16):
    print(k, sum(res[k])/len(res[k]),sep=': ')
```

```python
#optimal k is k=3 or 4!
```
