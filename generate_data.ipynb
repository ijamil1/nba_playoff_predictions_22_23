```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('data/data_v5.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0.1','Unnamed: 0', 'Year','t1_TEAM','t2_TEAM','t1_series_wins','t2_series_wins'],inplace=True)
```

```python
X = df.drop(columns=['outcome']).to_numpy()
y = df['outcome'].to_numpy()
x_train, x_test, y_train, y_test = train_test_split(X,y)
```

```python
def accuracy(y1,y2):
    cor = 0
    for i in range(y1.shape[0]):
        if y1[i]==y2[i]:
            cor+=1
    return cor/y1.shape[0]
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
ypreds = lr.predict(x_test)
test_res = accuracy(ypreds,y_test)
train_res = accuracy(lr.predict(x_train),y_train)
```

```python
print(train_res, test_res, sep=', ')
```

```python
#relatively poor performance on the data we see during training 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
rfc.fit(x_train,y_train)
ypreds = rfc.predict(x_test)
test_res = accuracy(ypreds,y_test)
train_res = accuracy(rfc.predict(x_train),y_train)
```

```python
print(train_res, test_res, sep=', ')
```

```python
#by construction, RFC is perfect on the data it sees during training
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
nb.fit(x_train,y_train)
print(accuracy(nb.predict(x_train),y_train),accuracy(nb.predict(x_test),y_test),sep=', ')
```

```python
#again, relatively poor performance on the data we see during training 
```

```python
#will generate synthetic data to augment data set; will hopefully enable more succesful learning
#idea: use Gaussian mixture models to generate data; how I'm thinking to do this
#break dataset into 2 subsets based on training label, for each subset, run GMM on the subset for some k, then
#generate more data from this dataset and assign it the corresponding label associated with the subset 
#generate data from the subsets in the same ratio as the labels appear in the original data
#Then, once we have the synthetic data, do some EDA to ensure data looks similar to real set

#use AIC score to determine proper k (number of Gaussians), does not need to be same k across subsets
```

```python
from sklearn.mixture import GaussianMixture as gmm
```

```python
df.head()
```

```python
one = df[df['outcome']==1]
zero = df[df['outcome']==0]
```

```python
one.head()
```

```python
one = one.drop(columns=['outcome']).to_numpy()
zero = zero.drop(columns=['outcome']).to_numpy()
```

```python
one_aic = []
zero_aic = []

for i in range(1,101):
    gmm1 = gmm(n_components=i,covariance_type='full',random_state=0).fit(one)
    gmm0 = gmm(n_components=i,covariance_type='full',random_state=0).fit(zero)
    one_aic.append(gmm1.aic(one))
    zero_aic.append(gmm0.aic(zero))
```

```python
np.argmin(zero_aic)
```

```python
np.argmin(one_aic)
```

```python
#so, for zero, k = 4 and for one, k = 2
```

```python
gmm1 = gmm(n_components=2,covariance_type='full',random_state=0)
gmm1.fit(one)

gmm0 = gmm(n_components=4,covariance_type='full',random_state=0)
gmm0.fit(zero)
```

```python
#determine what percent of original dataset is classified as 1 and 0
df[df['outcome']==1].shape[0]/df.shape[0]
```

```python
#52% is classifed as 1
one_samples = gmm1.sample(1000)[0] #just features
zero_samples = gmm0.sample(1000)[0]

one_samples = np.hstack((one_samples,np.ones((1000,1)))) #label appended
zero_samples = np.hstack((zero_samples,np.zeros((1000,1))))

syn_data_arr = np.vstack((one_samples,zero_samples))
print(syn_data_arr.shape)
```

```python
syn_data_arr[:,218]
```

```python
syn_df = pd.DataFrame(syn_data_arr,columns=df.columns)
syn_df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
import math

num_cols = len(df.columns)-1
num_test = math.floor(0.05*num_cols) #5% of the number of cols

#choose num_test cols from num_cols cols randomly
col_idxs = np.random.choice(np.arange(num_cols),size=num_test,replace=False)

for idx in col_idxs:
    orig = df.iloc[:,idx].to_numpy()
    new = syn_df.iloc[:,idx].to_numpy()
    
    plt.hist(orig,density=True,color='b',label='original')
    plt.hist(new,density=True,color='r',label='synthetic')
    plt.title('Comparing orig and syn data for ' + df.columns[idx])
    plt.legend()
    plt.savefig('col ' + str(idx))
    plt.close()

```

```python
#after visual inspection, the synthetic data looks pretty similar to the original data

union_df = pd.concat([df,syn_df])
union_df.to_csv('data/data_v7.cs')
```
