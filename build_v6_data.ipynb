```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/data_v5.csv')
df.head()
```

```python
df.drop(columns=['Unnamed: 0', 'Unnamed: 0.1','Year','t1_TEAM','t2_TEAM','t1_series_wins','t2_series_wins'],inplace=True)
```

```python
outcomes = df['outcome']
df = df.drop(columns=['outcome'])
```

```python
cols = list(df.columns)
suffixes = []
for col in cols:
    if 't1' in col:
        suffixes.append(col[col.find('t1_')+3:])
    elif 't2' in col:
        suffixes.append(col[col.find('t2_')+3:])
    else:
        print(col)
```

```python
suffixes
```

```python
d = {}
for suf in suffixes:
    ar = np.subtract(df['t1_'+suf].to_numpy().reshape(-1), df['t2_'+suf].to_numpy())
    d[suf+'_diff'] = ar[:]
v6 = pd.DataFrame(d)
```

```python
v6.head()
```

```python
v6['outcome']=outcomes
```

```python
v6.to_csv('data/data_v6.csv',index=False)
```
