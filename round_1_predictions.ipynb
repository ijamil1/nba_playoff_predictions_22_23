```python
coef_df = pd.read_csv('data/coefs_lr.csv',header=None)
coef_df.head()
```

```python
coefs_lr = coef_df.to_numpy()[0,:]
```

```python
int_df = pd.read_csv('data/intercept_lr.csv',header=None)
int_df.head()
```

```python
intercept = int_df.iloc[0]
intercept[0]
```

```python
def return_class(x):
    #expect np array for x
    denom = 1 + np.exp(-1*(np.dot(x,coefs_lr.reshape(-1))+intercept[0]))
    prob = 1/denom
    if prob > 0.5:
        return 1
    return 0
```

```python
y_preds = np.zeros(test.shape[0]).reshape(-1)
test = test.to_numpy()
for i in range(test.shape[0]):
    y_preds[i] = return_class(test[i,:].reshape(-1))
```

```python
for i in range(t1.shape[0]):
    if y_preds[i] == 1:
        print(t1.iloc[i], t2.iloc[i], sep=' beats ')
    else:
        print(t2.iloc[i], t1.iloc[i], sep=' beats ')
```
